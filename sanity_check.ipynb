```python
import spacy
from spacy.tokens import DocBin, Doc
```

```python
# load files in corpus/cdt/ to check for data leakage 
nlp = Doc(vocab=spacy.blank("da").vocab)
Doc.set_extension("doc_id", default=None)

files = ["/work/training_practice/ddt_dane_cdt_project/corpus/cdt/train.spacy",
         "/work/training_practice/ddt_dane_cdt_project/corpus/cdt/dev.spacy",
         "/work/training_practice/ddt_dane_cdt_project/corpus/cdt/test.spacy"]

doc_ids = {"train": [], "dev": [], "test": []}


for split, file in zip(["train", "dev", "test"], files):
    db = DocBin().from_disk(file)
    docs = list(db.get_docs(nlp.vocab))
    for doc in docs: 
        doc_ids[split].append(doc._.doc_id)
```

```python
for split, file in zip(["train", "dev", "test"], files):
    db = DocBin().from_disk(file)
    docs = list(db.get_docs(nlp.vocab))
    print(f"len of {split}:",len(docs))
```

```python
# check for data leakage (same doc_id across splits)
train = set(doc_ids["train"])
dev = set(doc_ids["dev"])
test = set(doc_ids["test"])

print("train and dev overlap:", train & dev)
print("train and test overlap:", train & test)
print("dev and test overlap:", dev & test)
```

```python
# check structure (mean n of sentences per doc)
for split, file in zip(["train", "dev", "test"], files):
    db = DocBin().from_disk(file)
    docs = list(db.get_docs(nlp.vocab))
    
    n_sents = []
    for doc in docs:
        n_sents.append(len(list(doc.sents)))
    print(str(split), "mean n of sentences per doc", round(sum(n_sents) / len(n_sents), 2))
    
# data grouped by about 10 sentences
```
